### Import required libraries

```python
%matplotlib inline
from collections import defaultdict
import json
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib as mpl
```

### Massage the dataset to merge state data files into one based on the state code file

```python
# Read the state code file
state_cols = ['st_name', 'st_abbv']
states = pd.read_csv('data/statecodes.csv', names=state_cols)

pieces = []
columns = ['state', 'sex', 'year', 'name', 'births']
for state in states['st_abbv']:
    path = 'data/%s.txt' % state
    frame = pd.read_csv(path, names=columns)
    frame['state'] = state
    pieces.append(frame)
    # Merge all read dataframes into one dataframe
    allNames = pd.concat(pieces, ignore_index=True)
```

##### The merged dataframe is very large. Only data from year 2000 and above will be used.

```python
# Data from 2000 and above
allNames = allNames[allNames.year >= 2000]
```

##### Adding the column prop, which is the proportion of babies given a certain name with respect to its state, year, and sex. 

```python
def Add_Prop_Col(group):
    births = group.births.astype(float)
    group['prop'] = births / births.sum()
    return group

allNames = allNames.groupby(['state', 'year', 'sex']).apply(Add_Prop_Col)
allNames
```

##### Taking the top 1000 names for each state/year/sex combination, then sorting them in descending order.

```python
pieces = []
for year, group in allNames.groupby(['state', 'year', 'sex']):
    pieces.append(group.sort_index(by='births', ascending=False)[:1000])
top1000 = pd.concat(pieces, ignore_index=True)
top1000
```

### Start the process of building the Map

##### The next cell defines a function that takes a JSON geography filename and returns a dict that maps every id to its geometry. For this application (using input file us-states.json), the ids are two-character state codes such as FL, GA, and NY.

```python
def regionsToGeometry(geofile):
    # Geofile is a json geography file. Returns a dict
    # mapping each feature id to a list of polygons 
    # describing its geography
    res = defaultdict(list)
    data = json.load(file(geofile))
    for f in data['features']:
        region = f['id']
        geo = f['geometry']
        if geo['type'] == 'Polygon':
            for coords in geo['coordinates']:
                res[region].append(coords)
        elif geo['type'] == 'MultiPolygon':
            for polygon in geo['coordinates']:
                res[region].extend(polygon)
    return res
```

##### Call the regionsToGeometry function using the regions json file:

```python
geofile = 'data/us-states.json'
regions = regionsToGeometry(geofile)
```

##### The function below takes this regions dict, a single id (region) in the key set of regions, and an axes. It draws the region's geometry into the axis.

```python
def drawRegion(regions, region, ax, **kwargs):
    for polygon in regions[region]:
        xs, ys = zip(*polygon)
        ax.fill(xs, ys, **kwargs)
    return ax
```

##### The below function takes regions dictionary and renders its geometry. The x and y ranges are set so that the full geometry is properly framed, and set_xlim and set_ylim are used for that purpose. Also the tickmarks are turned off.  The function below takes a regions dictionary and renders its geometry. The x and y ranges need to be set so that the full geometry is properly framed, using the set_xlim and set_ylim. Besides, the tickmarks are turned off.

```python
def drawRegions(regions, **kwargs):
    fig = plt.figure(figsize=(12,8))
    ax = plt.gca()
    ax.set_xticks([])
    ax.set_yticks([])
    # continental US x and y range
    ax.set_xlim(-130, -60)
    ax.set_ylim(22, 52)
    for region in regions.keys():
        drawRegion(regions, region, ax, **kwargs)
```

##### The graph generated showed that the states are assigned colors at random. The same map-drawing approach used in drawRegions, will be utilized to create choropleths in which colors are based on some measure. A dataframe is created first that captures the population of each state, using the same state codes.

```python
pop = pd.read_csv('data/statesByPopulation.csv', names=['state', 'pop'])
popDF = pop.set_index('state')
popDF
```

##### Define our choropleth function and call it with the regions dict, DataFrame data, and a column name col. The function uses the colormap cmap to map the values data[region][col] to colors, which are used to paint region.

```python
def makeChoroplethMap(regions, data, col, label=None, cmap=cm.Spectral, figsize=(12,8), **kwargs):
    fig = plt.figure()
    fig.set_figwidth(figsize[0])
    fig.set_figheight(figsize[1])
    ax = plt.gca()
    ax.set_position([0.0, 0.0, 0.94, 0.96])
    ax.set_xlim(-130, -60)
    ax.set_ylim(22, 52)
    knownStates = set(data.index)
    vmin = data[col].min()
    vmax = data[col].max()
    norm = mpl.colors.Normalize(vmin=vmin, vmax=vmax)
    for region in regions:
        if region in knownStates:
            color = cmap(norm(data.ix[region][col]))
            drawRegion(regions, region, ax, color=color, **kwargs)
    if label:
        ax.set_title(label, loc='center')
    ax1 = fig.add_axes([0.90, 0.0, 0.04, 0.96])
    cb1=mpl.colorbar.ColorbarBase(ax1, cmap=cmap,
                                  norm=norm,
                                  orientation='vertical')
    ax.set_xticks([])
    ax.set_yticks([])
    fig.canvas.draw()
    return ax
```

### The makeMap() Function

##### After the preparation done above, the function to draw the map is next. As input, the function to draw the map requires the regions dictionary, top1000 DataFrame as a dataset, sex, quantile value, and other 'map drawing' arguments. Based on year, sex, and quantile, the function processes the data into a new DataFrame, generates a label, and finally renders the map.

```python
def get_quantile_count(group, q=0.5): 
    group = group.sort_index(by='prop', ascending=False)
    return group.prop.cumsum().searchsorted(q)+1 

yr2002 = top1000[top1000.year == 2002]
div2002 = yr2002.groupby(['state', 'sex']).apply(get_quantile_count) 
div2002 = div2002.unstack('sex') 
makeChoroplethMap(regions, div2002, 'M', label='Diversity (50%): 2002, Male')
```

```python
def makeLabel(charSex, q, year):
    if (charSex == 'M'):
        strSex = 'Male'
    elif (charSex == 'F'):
        strSex = 'Female'
    else:
        # Default to Male if the entry is invalid
        strSex = 'Male'
        
    strLabel = "Diversity ({quantile}%): {theYear}, {theSex}".format(quantile=int(q*100),theYear=year,theSex=strSex)
    return strLabel
```

```python
def makeMap(regions, data, year, sex, q=0.5, cmap=cm.Spectral, **kwargs):
    # Build the new DataFrame
    rdata = data[data.year == year]
    rdata = rdata.groupby(['state', 'sex']).apply(get_quantile_count,q) 
    rdata = rdata.unstack('sex') 
    
    # Generate the label
    lbl = makeLabel(sex, q, year)
    
    # Draw the map
    makeChoroplethMap(regions, rdata, sex, label=lbl, cmap=cmap, **kwargs)
    
makeMap(regions, top1000, 2002, 'M', q=0.5)
```
